```python
from Grid import Grid
from Cell import Cell
import numpy as np
import matplotlib.pyplot as plt
from Patterns import Pattern, AVAILABLE_PATTERNS
from utils import compute_entropy
```

```python
grid = Grid(
    length=128, 
    width=128, 
    seed=42, 
    pattern=None,)
```

```python
entropy_evolution = []
```

```python
from IPython.display import clear_output
fig, ax = plt.subplots(figsize=(4, 4))

for _ in range(200):
    grid.update_grid()
    entropy_evolution.append(compute_entropy(grid))
    clear_output(wait=True)
    ax.clear()
    ax.imshow(grid.get_grid(), cmap='binary')
    ax.set_title(f'Game of Life')
    plt.pause(1e-5)
    display(fig)
```

```python
plt.figure()
plt.plot(entropy_evolution)
# plot mean entropy line of the last 100 generations
mean_entropy = np.mean(entropy_evolution[-200:])
plt.axhline(y=mean_entropy, color='r', linestyle='--', label='Mean Entropy (last 200 gens)')
plt.legend()
plt.title('Entropy Evolution')
plt.xlabel('Generation')
plt.ylabel('Entropy')
plt.show()
```
